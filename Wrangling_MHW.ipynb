1. Read the abstract. What is this paper about?

The paper by Hadley Wickham talks about making data easier to work with, which he believes is understudied. He introduces the idea of "tidy data," which means organizing your data consistently so that each column has one type of information, each row is one record, and similar information is kept together. The paper explores the benefits of organzing data this way

2. Read the introduction. What is the "tidy data standard" intended to accomplish?

The tidy data standard lays out a standard format in which data can be organized so that data scientists can spend more time interpreting the data rather than dealing with the logistics of organizing data.

2. Categorical variable: For the ./data/sharks.csv data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.

To clean this, I want to create a line of code that drops unhelpful values such as invalid and questionable, as well as combine types that are of similar vein, such as "boating" "boat" and "boatomg"

```python
type=sharks['Type']
type= type.replace(["Sea Disaster", "Watercraft", "Boat", "Boatomg"], "Boating")
type.value_counts()

type= type.replace(["Questionable", "Invalid", "Unconfirmed", "Unverified", "Under investigation"], np.nan)
type.value_counts()

df['Type'] = type
del type
df['Type'].value_counts()
```

4. Missing values, not at random: For the pretrial data covered in the lecture, clean the ImposedSentenceAllChargeInContactEvent variable as well as you can, and explain the choices you make. (Hint: Look at the SentenceTypeAllChargesAtConvictionInContactEvent variable.)

```python
df = pd.read_csv('pretrial2017.csv', low_memory=False)
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

length = pd.to_numeric(length,errors='coerce')
length_NA = length.isnull()
print( np.sum(length_NA),'\n') #

```

there are a significant amount of missing values, so there needs to be more manipulation

```python
print( pd.crosstab(length_NA, type), '\n')

length = length.mask( type == 4, 0)
length = length.mask( type == 9, np.nan)

length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n')

df['ImposedSentenceAllChargeInContactEvent'] = length
del length, type
```

^^These lines of code, first prints a crosstab (contingency table) of missing values in length_NA against the type column. It then modifies the length column by replacing values where type == 4 with 0, and where type == 9 with NaN. this reduces the number of missing variables. After that, it recalculates length_NA to check which values in length are missing (NaN), prints the updated crosstab, and the total number of missing values. Finally, the column length is assigned to df['ImposedSentenceAllChargeInContactEvent'], and the length and type variables are deleted to free up memory.
