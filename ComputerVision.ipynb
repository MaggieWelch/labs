4. Use the reshaped X_test and y_test data to create a k- nearest neighbor classifier of digit. What is the optimal number of neighbors
k? If you can't determine this, play around with different values of for your classifier.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Use a smaller subset to avoid crashing during testing
X_train_small = X_train_reshaped[:6000]
y_train_small = y_train[:6000]
X_test_small = X_test_reshaped[:1000]
y_test_small = y_test[:1000]

# Define a range of k values to test
k_values = range(1, 20)  # You can change the range to test more or fewer values of k

# Store accuracy results for each k
accuracy_results = []

# Loop over different values of k to find the optimal one
for k in k_values:
    # Initialize the kNN classifier with the current k value
    knn = KNeighborsClassifier(n_neighbors=k, algorithm='kd_tree')

    # Train the classifier on the smaller training set
    knn.fit(X_train_small, y_train_small)

    # Predict the labels for the test set
    y_pred = knn.predict(X_test_small)

    # Calculate accuracy for this value of k
    accuracy = accuracy_score(y_test_small, y_pred)

    # Append the accuracy result for this k
    accuracy_results.append((k, accuracy))

# Find the k with the highest accuracy
optimal_k = max(accuracy_results, key=lambda item: item[1])[0]

# Print the optimal k and the accuracy associated with it
print(f"The optimal number of neighbors (k) is: {optimal_k}")

# Train a final model using the optimal k and evaluate its performance on the test set
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k, algorithm='kd_tree')
knn_optimal.fit(X_train_small, y_train_small)

# Make predictions using the test set
y_pred_optimal = knn_optimal.predict(X_test_small)

# Calculate and print the accuracy using the optimal k
optimal_accuracy = accuracy_score(y_test_small, y_pred_optimal)
print(f"Accuracy with the optimal k ({optimal_k}) on the test set: {optimal_accuracy:.4f}")

# Optional: print out accuracy results for all k values
for k, accuracy in accuracy_results:
    print(f"k = {k}, Accuracy = {accuracy:.4f}")
```
